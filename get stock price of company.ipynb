```python
import pandas_datareader.data as web
import datetime
import copy
```

```python
#start從2018/01/01開始，多取兩週時間，以便抓出上一個交易日data
start = datetime.date(2020, 10, 14)
start_delta = start - datetime.timedelta(weeks=2) #多取兩個禮拜的data
end = datetime.date(2020, 10, 22)#datetime.date.today()
```

# APPLE

```python
AAPL = web.DataReader('AAPL', 'stooq',start_delta, end)
#drop unnecessary column & reverse index
apple = AAPL.drop(["Open", "High", "Low", "Volume"], axis=1).reindex(index=AAPL.index[::-1])
apple['daily_return'] = ((apple['Close']/apple['Close'].shift(1))-1) * 100
# apple
```

```python
# copy data & insert new column
apple_price = copy.deepcopy(apple)
apple_price.insert(loc=2, column='UpDown', value='') 

apple_price.loc[apple_price['daily_return']>0, 'UpDown'] = 1
apple_price.loc[apple_price['daily_return']==0, 'UpDown'] = 0
apple_price.loc[apple_price['daily_return']<0, 'UpDown'] = -1

# 切割 -> 判別2018-01-01前一天工作日
for i in range(apple_price.size):
    if apple_price.index[i] >= start:
        apple_price = apple_price[i:]
        break
apple_price
```

```python
company_symbol = ['AMZN', 'AAPL', 'FB', 'GOOGL', 'MSFT']
company_symbol
```

```python
for name in company_symbol:
    data = web.DataReader(name, 'stooq',start_delta, end)
    #drop unnecessary column & reverse index
    data = data.drop(["Open", "High", "Low", "Volume"], axis=1).reindex(index=data.index[::-1])
    data['daily_return'] = ((data['Close']/data['Close'].shift(1))-1) * 100
    
    data.insert(loc=2, column='UpDown',value='')#insert new column

    data.loc[data['daily_return'] > 2, 'UpDown'] = 1
    data.loc[data['daily_return'] < -2, 'UpDown'] = -1
    data['UpDown'] = data['UpDown'].replace('', 0)

    for i in range(data.size):
        if data.index[i] >= start:
            data = data[i:]
            break
    data['date'] = data.index#保留data，儲存成csv時index會並成0~n        
    data.to_csv(name + '_stock_price' + '.csv',index = False)
data
```
